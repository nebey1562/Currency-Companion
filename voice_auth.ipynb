```python
import os
import librosa
import numpy as np
import pickle
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score
# Define dataset directory
DATASET_DIR = "/Users/jamesjoseph/Jamezzz/Final project/voice-auth/differentPhrase"

# Function to extract MFCC features from audio
def extract_features(file_path, n_mfcc=13):
    audio, sr = librosa.load(file_path, sr=None)  # Load audio file
    mfccs = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=n_mfcc)  # Extract MFCCs
    mfccs_mean = np.mean(mfccs.T, axis=0)  # Take mean across time axis
    return mfccs_mean
# Prepare data and labels
X = []  # Features
y = []  # Labels (speaker names)

# Iterate through the dataset directory
for speaker in os.listdir(DATASET_DIR):
    speaker_dir = os.path.join(DATASET_DIR, speaker)
    if os.path.isdir(speaker_dir):
        for file in os.listdir(speaker_dir):
            if file.endswith(".flac"):
                file_path = os.path.join(speaker_dir, file)
                features = extract_features(file_path)
                X.append(features)
                y.append(speaker)  # Speaker's name as the label

# Convert to numpy arrays
X = np.array(X)
y = np.array(y)

# Encode labels into integers
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Save the label encoder for future use
with open("label_encoder.pkl", "wb") as le_file:
    pickle.dump(label_encoder, le_file)

print(f"Extracted features from {len(X)} audio files.")
print(len(y_encoded))
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Train an SVM model
model = SVC(kernel="linear", probability=True)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)  # Ensure that only X_test is used here
# Save the trained model
with open("speaker_recognition_model.pkl", "wb") as model_file:
    pickle.dump(model, model_file)

print("Model trained and saved.")
# Ensure labels correspond to the unique classes in y_test
labels = np.unique(y_test)
print("y_pred shape:", y_pred.shape)
# Update the classification report
print("Classification Report:")
print(classification_report(y_test, y_pred, labels=labels, target_names=label_encoder.classes_))
print(f"Accuracy: {accuracy_score(y_test, y_pred) * 100:.2f}%")

```
