```python
import tensorflow as tf
model = tf.keras.models.load_model('currency_detection_model.h5')
print("Model Loaded!")
```

```python
import cv2
import numpy as np
cap = cv2.VideoCapture(0)  

class_labels = ['10', '20', '50', '100', '200', '500']

def predict_currency(image):
    """Preprocess and predict the currency denomination of an image."""
    # Resize and preprocess the image
    resized_image = cv2.resize(image, (224, 224)) / 255.0  # Normalize
    image_expanded = np.expand_dims(resized_image, axis=0)  # Add batch dimension

    predictions = model.predict(image_expanded)
    predicted_class = np.argmax(predictions)
    return class_labels[predicted_class]

while True:
   
    ret, frame = cap.read()
    if not ret:
        print("Failed to capture frame. Exiting...")
        break
    cv2.imshow("Currency Detection - Live Feed", frame) # Display the live feed
    key = cv2.waitKey(1) & 0xFF # Check for user input

    if key == ord('c'):
        print("Capturing frame...")
        captured_frame = frame.copy()
        cv2.imshow("Captured Frame", captured_frame)

        predicted_currency = predict_currency(captured_frame)

        cv2.putText(captured_frame, f"Predicted: {predicted_currency}", (10, 30),
                    cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
        cv2.imshow("Captured Frame with Prediction", captured_frame)

        print(f"Predicted Currency: {predicted_currency}")

    if key == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
