**Resample the audio**

```python
import soundfile as sf
from scipy.signal import resample

def resample_audio(input_path, output_path, target_rate=16000):
    speech, rate = sf.read(input_path)
    if rate != target_rate:
        print(f"Resampling audio from {rate} Hz to {target_rate} Hz...")
        duration = len(speech) / rate
        target_length = int(duration * target_rate)
        resampled_speech = resample(speech, target_length)
        sf.write(output_path, resampled_speech, target_rate)
        print(f"Audio resampled and saved to {output_path}")
    else:
        print(f"Audio is already at {target_rate} Hz. No resampling needed.")
        sf.write(output_path, speech, rate)
    return output_path
```

**Speech to Text**

```python
from transformers import Wav2Vec2ForCTC, Wav2Vec2Processor
import torch
import soundfile as sf

# Load model and tokenizer
processor = Wav2Vec2Processor.from_pretrained("facebook/wav2vec2-base-960h")
model = Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-base-960h")

def transcribe_with_wav2vec(audio_file_path):

        speech, rate = sf.read(audio_file_path)
        input_values = processor(speech, sampling_rate=rate, return_tensors="pt").input_values
        logits = model(input_values).logits
        predicted_ids = torch.argmax(logits, dim=-1)
        transcription = processor.decode(predicted_ids[0])
        print("Transcription:", transcription)
        return transcription.lower()
```

**Text to Speech**

```python
from gtts import gTTS
import os
def text_to_speech(text, filename):
    text=text
    language = 'en'

    tts = gTTS(text=text, lang=language, slow=False)
    file_path = f"/content/drive/MyDrive/TTS_Output/{filename}.wav"
    tts.save(file_path)
    return tts

```

**Handle the command**

```python
def handle_command(command):
    if command is None:
        response = "I couldn't understand your command. Please try again."

        print(response)
        text_to_speech(response,"none")
        return response

    if "redirect" in command or "redirects" in command:
        response = "Redirection successful"
    elif "transfer" in command or "pay" in command or "payment" in command:
        response = "Confirm transaction?"
        print(response)
        text_to_speech(response, "confirm")

        co=resample_audio("/content/drive/My Drive/TTS_Output/output1.wav", "/content/drive/My Drive/TTS_Output/output1_resampled.wav")
        confirmation = transcribe_with_wav2vec(co)

        if confirmation == "yes":
            response = "Transaction successful"
        else:
            response = "Transaction failed"
    else:
        response = "Command not recognized. Please try again."

    text_to_speech(response, "final_response")
    return response
```

```python

input_audio_file = "/content/drive/My Drive/TTS_Output/output2.wav"
resampled_audio_file = "/content/drive/My Drive/TTS_Output/output2_resampled.wav"
resampled_audio_path = resample_audio(input_audio_file, resampled_audio_file)
transcription = transcribe_with_wav2vec(resampled_audio_path)
handle_command(transcription)
```
